```python
%pip install numpy
%pip install matplotlib scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
```

```python
# Lennard-Jones potential parameters
sigma = 1.0  # Characteristic distance
epsilon = 1.0  # Depth of potential well
```

```python
def lennard_jones_potential(r, sigma, epsilon):
    return 4 * epsilon * ((sigma / r) ** 12 - (sigma / r) ** 6)

def total_energy(positions):
    distances = pdist(positions)
    lj_potential = lennard_jones_potential(distances, sigma, epsilon)
    return np.sum(lj_potential)

def initialize_positions(n_particles, box_size):
    return np.random.rand(n_particles, 3) * box_size
```

```python
# Simulation parameters
n_particles = 10
box_size = 10.0
positions = initialize_positions(n_particles, box_size)

# Compute energy
total_E = total_energy(positions)
print(f"Total Energy of the system: {total_E}")
```

```python
# Plot the system
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(positions[:, 0], positions[:, 1], positions[:, 2], c='b', marker='o')
ax.set_xlim(0, box_size)
ax.set_ylim(0, box_size)
ax.set_zlim(0, box_size)
ax.set_xlabel("X position")
ax.set_ylabel("Y position")
ax.set_zlabel("Z position")
ax.set_title("Initial Molecular Configuration")
plt.show()
```
